```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# path = '/home/jichao/Temp/test3grade.csv'
gradebook = pd.read_csv('test3grade.csv')
gradebook.head()
```

```python
def findtestname(keyward, column):
    for str in column:
        if keyward in str:
            return str
    else:
        print(f"Cannot find {keyward} in {column}")
        return None
# Find the name of the column
    
testname = findtestname('Test3', gradebook.columns)
print(testname)
```

```python
test3grade = pd.to_numeric(gradebook[testname], errors='coerce')
test3grade.hist(bins=10)
```
